# Computing time  vs n for (c2 vs c4_1 vs c4_2)

```python
from c2 import c2
from c4_1 import c4_1
from c4_2 import c4_2
import time

c2_times = []
c4_1_times = []
c4_2_times = []
MAX_N = 25
for fun, times in [(c2,c2_times), (c4_1, c4_1_times),(c4_2, c4_2_times)]:
    print(fun.__name__)
    for n in range(1,MAX_N+1):
        if n!=MAX_N:
            print('\t'+str(n) + '/'+ str(MAX_N), end="\r")
        else:
            print('\t'+str(n) + '/'+ str(MAX_N))
        start = time.time()
        fun(n)
        end = time.time()
        times.append(end-start)
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
ax = plt.figure().gca()
fig = plt.gcf()
ax.set_yscale('log')
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.set_xticks(range(1,len(c2_times)+1))
plt.plot(range(1,len(c4_1_times)+1), c4_1_times, label='c4_1')
plt.plot(range(1,len(c4_2_times)+1), c4_2_times, label='c4_2')
plt.plot(range(1,len(c2_times)+1), c2_times, label='c2')
plt.xlabel('n')
plt.ylabel('time')
plt.suptitle('Comparition of Time Performance of c2, c4_1 and c4_2')
fig.set_size_inches(15,6)
ax.legend()
plt.show()
```

# Computing time of c5_1 vs c6_1

```python
from c5 import c5
from c6 import c6
import time

c5_1_times = []
c6_1_times = []
data_folder_path = 'optpr1-20191011'
MAX_N = 25
for fun, times in [(c5,c5_1_times), (c6, c6_1_times)]:
    print(fun.__name__)
    for n in range(1,MAX_N+1):
        if n!=MAX_N:
            print('\t'+str(n) + '/'+ str(MAX_N), end="\r")
        else:
            print('\t'+str(n) + '/'+ str(MAX_N))
        start = time.time()
        fun(data_folder_path, n=100)
        end = time.time()
        times.append(1000*(end-start))
```

```python
print('problem: ', data_folder_path)
print('c5 time:', str(round(sum(c5_1_times) / len(c5_1_times),2))+'ms')
print('c6 time:', str(round(sum(c6_1_times) / len(c6_1_times),2))+'ms')
```
